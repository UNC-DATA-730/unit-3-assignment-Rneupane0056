## Linear model with one numerical variable

Execute the following cell to load the [ISLR2](https://cran.rstudio.com/web/packages/ISLR2/index.html) and [Tidverse](https://www.tidyverse.org/) packages.

````r
library('tidyverse')
library('ISLR2')
````

The `ISLR2` package provides a dataset called `Boston` that we will use in this assignment.

````r
Boston |> glimpse()
````

üö® Use the **Jupyter Lab Contextual Help Feature** to see the documentation for this data. 

‚ùì **In the markdown cell below, add definitions for the `lstat` and `medv` variables in the `Boston` data.** 

- `lstat`: This is a double that shows the percentage of the lower status of the population.
- `medv`: This is a double that shows the median value of owner-occupied homes in $1000.

Let's plot `medv` (y-axis) versus `lstat`.

````r
p = ggplot(data = Boston, aes(x = lstat, y = medv)) + 
    geom_point() +
    labs(x = "% Lower Status of Population",
         y = "Median Home Value ($1000)",
         title = "Median Home Value (medv) vs. Lower Status Population (lstat)")
p
````

‚ùìDoes the relationship appear to be positive or negative? Does it look to be reasonably linear?

### **Answer:** The relationship appears to be negative with a reasonably linear slope. There appears to be a negative relationship between the Median Home Value ($1000) when the % Lower Status of Population increases. 

### Let's build a model!

We need to first load the [Tidymodels]() package.

````r
library('tidymodels')
````

First we specify our model as `linear regression` using the `lm` engine.

````r
mod = linear_reg() |> set_engine("lm")

mod
````

Next we "fit" our model by supplying the `formula` and the data.

````r
mod_fit = mod |> fit(medv ~ lstat, data = Boston)

mod_fit
````

### Get the regression table

````r
# We can use the tidy function to get a table of our model information
tidy(mod_fit)
````

‚ùìComparing on the chart above and the regression table, does the **sign** of the `estimate` for the `lstat` term coefficient fit your expectations?

### **Answer:** Yes, the negative sign of -0.95 for lstat makes sense and matches my expectations. The scatter plot showed that as lstat increased along the x-axis, the medv deceased along the y-axis. With a p-value of 5.08 x 10^-88, an extremely small value, this negative relationship is highly statistically significant and not likely due to chance.

### Making predictions and visualizing the model

We can use the `augment` function to "predict" `medv` for all the values in our original dataset. We will capture these predictions in a new data frame called `Boston2`. The predicted values are found in the `.pred` column.

````r
Boston2 = augment(mod_fit, Boston)

glimpse(Boston2)
````

Let's visualize our model.

````r
p = ggplot(Boston2, aes(x = lstat)) +
    geom_point(aes(y = medv)) + 
    geom_line(aes(y = .pred), color = 'coral', linewidth = 1.5)
p
````

```r
# The code below assumes your predictions column is called ".pred" and is in a

ggplot(Carseats2, aes(x = ShelveLoc)) +
```

````r
# Fit the model with ShelveLoc as predictor
mod_shelf = linear_reg() |> 
  set_engine("lm") |> 
  fit(Sales ~ ShelveLoc, data = Carseats)

# Show the regression table
tidy(mod_shelf)
````

````r
# Generate predictions and bind to original data
Carseats2 = mod_shelf |> 
  augment(new_data = Carseats)

# View the first few rows to confirm .pred column exists
head(Carseats2)
````

````r
# Create visualization with jittered points and predicted means
ggplot(Carseats2, aes(x = ShelveLoc)) + 
  geom_point(aes(y = Sales), 
             na.rm = TRUE, 
             position = position_jitter(height = 0, width = 0.2, seed = 42),
             alpha = 0.5) +
  geom_crossbar(aes(y = .pred, ymin = .pred, ymax = .pred), 
                color = 'coral',
                width = 0.5) +
  labs(
    x = "Shelf Location",
    y = "Sales (in thousands)",
    title = "Car Seat Sales by Shelf Location",
    subtitle = "Coral bars show predicted mean sales for each location"
  ) +
  theme_minimal()
````

Which `ShelveLoc` category does your intercept term represent? 

## **Answer:** 
### The "Bad" shelf location category is represented by the intercept term. Since there are three options for the Shelveloc category, the parameters for the model are automatically assigned with the "Bad" option chosen as the intercept (or the base case) because it is the first alphabetical option. The intercept is the average sales for the "Bad" shelf locations. 
### We can also say, ShelveLocGood: shows how much higher (or lower) sales are for "Good" locations compared to "Bad", and ShelveLocMedium: shows how much higher (or lower) sales are for "Medium" locations compared to "Bad".
